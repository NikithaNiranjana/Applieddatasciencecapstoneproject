Analyzing Proximity to the Equator

```python

# Function to calculate distance to the Equator
def distance_to_equator(lat):
    equator = (0, 0)
    site = (lat, 0)
    return geopy.distance.distance(site, equator).km

# Calculate the distance of each launch site to the Equator
launch_sites_df['Distance to Equator (km)'] = launch_sites_df['Lat'].apply(distance_to_equator)
launch_sites_df[['Launch Site', 'Lat', 'Distance to Equator (km)']]
```

Analyzing Proximity to the Coast

```python
# Define some known coastal coordinates near the launch sites (approximations)
coast_coords = [
    (28.562302, -80.577356),  # CCAFS LC-40 near the Atlantic Ocean
    (28.563197, -80.576820),  # CCAFS SLC-40 near the Atlantic Ocean
    (28.573255, -80.646895),  # KSC LC-39A near the Atlantic Ocean
    (34.632834, -120.610745)  # VAFB SLC-4E near the Pacific Ocean
]

# Function to calculate the closest distance to the coast
def distance_to_closest_coast(lat, long, coast_coords):
    site = (lat, long)
    return min([geopy.distance.distance(site, coast).km for coast in coast_coords])

# Calculate the distance of each launch site to the closest coast
launch_sites_df['Distance to Coast (km)'] = launch_sites_df.apply(
    lambda row: distance_to_closest_coast(row['Lat'], row['Long'], coast_coords), axis=1
)
launch_sites_df[['Launch Site', 'Lat', 'Long', 'Distance to Coast (km)']]
```

Findings
Proximity to the Equator:

Most launch sites are not extremely close to the Equator but are located at latitudes that are relatively low (around 28 to 34 degrees north).
The proximity to the Equator is advantageous for launch sites because it provides a velocity boost due to Earth's rotation, making it more efficient to launch into geostationary orbits.

Proximity to the Coast:

All the launch sites are very close to the coast. This is essential because launching rockets over the ocean minimizes the risk to human populations in case of launch failures.
The calculated distances to the coast for all launch sites are quite small, confirming their close proximity to the ocean.

```python
# TASK 3: Calculate the distances between a launch site to its proximities
# Function to calculate distance between two geographical points
def calculate_distance(lat1, lon1, lat2, lon2):
    coords_1 = (lat1, lon1)
    coords_2 = (lat2, lon2)
    return geopy.distance.distance(coords_1, coords_2).km


distances_list = []
# Create a DataFrame to store distances
distances_df = pd.DataFrame(columns=['From Site', 'To Site', 'Distance (km)'])

# Calculate distances between each pair of launch sites
for i, row1 in launch_sites_df.iterrows():
    for j, row2 in launch_sites_df.iterrows():
        if i != j:
            distance = calculate_distance(row1['Lat'], row1['Long'], row2['Lat'], row2['Long'])
            distances_df = distances_df.append({
                'From Site': row1['Launch Site'],
                'To Site': row2['Launch Site'],
                'Distance (km)': distance
            }, ignore_index=True)

distances_df

```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
launch_site_name = 'CCAFS SLC-40'  # Example launch site name
launch_site_lat = launch_sites_df.loc[launch_sites_df['Launch Site'] == launch_site_name, 'Lat'].values[0]
launch_site_lon = launch_sites_df.loc[launch_sites_df['Launch Site'] == launch_site_name, 'Long'].values[0]
print(f"Launch Site: {launch_site_name}")
print(f"Latitude: {launch_site_lat}")
print(f"Longitude: {launch_site_lon}")
```
